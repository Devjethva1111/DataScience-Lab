```python
# 2. Draw a bar plot to show the number of persons with/without CHD and family history.
sns.countplot(x="famhist", hue="chd", data=df)
plt.xlabel("Family History of CHD")
plt.ylabel("Count")
plt.title("CHD and Family History")
plt.show()
```

```python
# 3. Does age have any correlation with sbp? Use a scatter plot.
sns.scatterplot(x="age", y="sbp", data=df) 
plt.xlabel("Age")
plt.ylabel("Systolic Blood Pressure (sbp)")
plt.title("Correlation between Age and sbp")
plt.show()
```

```python
# sns.heatmap(df.corr())

# Calculate the correlation for numeric columns
numeric_columns = df.select_dtypes(include=[float, int])  # Filter for numeric data types
correlation_matrix = numeric_columns.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 4.Compare the distribution of tobacco consumption for persons having CHD and not having CHD.Can you interpret the 
#   effect of tobacco consumption on having coronary heart disease?

sns.boxplot(x="chd", y="tobacco", data=df)
plt.xlabel("CHD")
plt.ylabel("Tobacco Consumption")
plt.title("Tobacco Consumption and CHD")
plt.show()
# Interpretation: The box plot shows that individuals with CHD tend to have higher tobacco consumption on average.
```

```python
# 5. How are the parameters sbp, obesity, age and ldl correlated? Choose the right plot to show the relationships.

sns.pairplot(df[["sbp", "obesity", "age", "ldl"]])
plt.show()
```

```python
#6. Derive a new column called agegroup from age column where persons falling in different age ranges
#    are categorized as:(0â€“15): young, (15â€“35): adults, (35â€“55): mid, (55â€“): old

# 6. Derive a new column called "agegroup" from "age".
def categorize_age(age):
    if age <= 15:
        return "young"
    elif 15 < age <= 35:
        return "adults"
    elif 35 < age <= 55:
        return "mid"
    else:
        return "old"

df["agegroup"] = df["age"].apply(categorize_age)
```

```python
df["agegroup"]
```

```python
# 7. Find out the number of CHD cases in different age categories. Do a barplot and sort them in the order of age groups.

chd_age_counts = df[df["chd"] == 1]["agegroup"].value_counts()
chd_age_counts = chd_age_counts.reindex(["young", "adults", "mid", "old"])  # Reorder the age groups

sns.barplot(x=chd_age_counts.index, y=chd_age_counts.values)
plt.xlabel("Age Group")
plt.ylabel("Number of CHD Cases")
plt.title("CHD Cases in Different Age Categories")
plt.show()
```

```python
# 8. Draw a box plot to compare distributions of ldl for differentÂ ageÂ groups.

sns.boxplot(x="agegroup", y="ldl", data=df, order=["young", "adults", "mid", "old"])
plt.xlabel("Age Group")
plt.ylabel("LDL Cholesterol")
plt.title("Distribution of LDL Cholesterol by Age Group")
plt.show()
```
