Ans 5:- Which month of the year typically sees most releases of high budgeted movies, that is,movies with budget of 25 crore or more?

Ans 6:-which are the top 10 movies with maximum return on investment(ROI)? Calculate return on investment(ROI) as(BoxOfficeCollection - Budget) / Budget

```python
df["ROI"]= (df["BoxOfficeCollection"]-df["Budget"]) / df["Budget"] 
```

```python
df[["MovieName","ROI"]].sort_values("ROI",ascending=False)[0:10] 
```

Ans 8:- Draw a histogram and a distribution plot to find the distribution of movie budgets.Interpret the plot to conclude if the most movies are high or low budgeted movies.

```python
# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Budget'], bins=30, kde=False, color='skyblue')
plt.title('Distribution of Movie Budgets')
plt.xlabel('Budget (Crore)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Plot distribution plot
plt.figure(figsize=(10, 6))
sns.histplot(df['Budget'], kde=True, color='skyblue')
plt.title('Distribution of Movie Budgets with KDE')
plt.xlabel('Budget (Crore)')
plt.ylabel('Density')
plt.show()
```
