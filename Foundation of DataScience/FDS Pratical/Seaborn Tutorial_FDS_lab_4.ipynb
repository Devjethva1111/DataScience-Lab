# Statistical analysis and visualization of airline-safety dataset

```python
# import all the libraries

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data set of airline-safety

df1= pd.read_csv("airline-safety.csv")
df1
```

```python
plt.figure(figsize=(14, 6))  # Adjust the figure size as needed
sns.boxplot(data=df1, palette="Set3")  # Choose an appropriate palette
plt.title("Distribution of Data")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
df1.isnull().sum()
```

```python
df_min_max_scaled = df1.iloc[:,1:]
  
# apply normalization techniques
for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())
```

```python
df_min_max_scaled.head()
```

```python
plt.figure(figsize=(14, 6))  # Adjust the figure size as needed
sns.boxplot(data=df_min_max_scaled, palette="Set3")  # Choose an appropriate palette
plt.title("Distribution of Data")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
df_min_max_scaled.describe()
```

```python
plt.figure(figsize=(14, 6))  # Adjust the figure size as needed
sns.boxplot(data=df_min_max_scaled, palette="Set3")  # Choose an appropriate palette
plt.title("Distribution of Data")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
df = pd.DataFrame(df_min_max_scaled)
```

```python
# Calculate quartiles and IQR
Q1 = df['avail_seat_km_per_week'].quantile(0.25)
Q3 = df['avail_seat_km_per_week'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Check for outliers
outliers = df[(df['avail_seat_km_per_week'] < lower_bound) | (df['avail_seat_km_per_week'] > upper_bound)]
```

```python
print("Outliers:")
print(outliers)
plt.boxplot(outliers)
plt.show()
```
