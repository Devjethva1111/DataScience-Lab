```python
import seaborn as sns
```

```python
df = pd.read_csv(r"H:\DataScience-Lab\Sem-2\Time series analysis & forecasting\files\AusAntidiabeticDrug.csv",parse_dates=['ds'])
df
```

```python
df["ds"] = pd.to_datetime(df["ds"])
print('Shape :',df.shape)
df.head()
```

```python
df["year"] = df["ds"].dt.year
df["month"] = df["ds"].dt.strftime('%b')
df.head()
```

```python
plt.figure(figsize=(16,7))
sns.pointplot(x="month",y="y", hue="year", data=df,order=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec'])
plt.xlabel("month")
plt.ylabel("$ millions")
plt.title("AntiDiabetic Drug Sales in Australia Segmented on Year")
plt.legend(loc='upper right')
```
