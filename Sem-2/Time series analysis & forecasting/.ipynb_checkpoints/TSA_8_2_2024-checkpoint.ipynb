```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"H:\DataScience-Lab\Sem-2\Time series analysis & forecasting\files\Monthly Steam Usage and Average Ambient Temperature.csv")
df
```

```python
# Extract features (temperature) and target variable (usage)
X = np.column_stack((np.ones(len(df)), df['Temperature']))
y = df['Usage'].values

# Calculate coefficients using the normal equation
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Print the coefficients
intercept, slope = theta
print(f'Intercept (theta_0): {intercept:.4f}')
print(f'Slope (theta_1): {slope:.4f}')
```

```python
import matplotlib.pyplot as plt

# Plot the data points
plt.scatter(df['Temperature'], df['Usage'], label='Data')

# Plot the regression line
regression_line = intercept + slope * df['Temperature']
plt.plot(df['Temperature'], regression_line, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('Temperature (Â°F)')
plt.ylabel('Usage/1000')
plt.title('Linear Regression Model')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
