```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
# Plot the autocorrelation function
plot_acf(df['Southern'], lags=100)  # Adjust lags as needed
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function of Southern Oscillation Data')
plt.show()
```

```python
# Create the autocorrelation plot
plt.figure(figsize=(10, 6))
pd.plotting.autocorrelation_plot(df2)
plt.title('Autocorrelation Plot')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()
```

```python
# Plot Partial Autocorrelation
plot_pacf(df['Southern'], alpha=0.05, use_vlines=True)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Plot')
plt.show()
```

```python
def autocorr(x):
    n = len(x)
    mean = np.mean(x)
    corr = np.correlate(x - mean, x - mean, mode='full')
    corr /= np.max(corr)
    return corr[n - 1:]


# Compute the autocorrelation function
corr = autocorr(df2["Y"])

# Plot the autocorrelation function
plt.plot(corr)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

```python
def autocorr(x):
    n = len(x)
    mean = np.mean(x)
    corr = np.correlate(x - mean, x - mean, mode='full')
    corr /= np.max(corr)
    return corr[n - 1:]


# Compute the autocorrelation function
corr = autocorr(df["Southern"])

# Plot the autocorrelation function
plt.plot(corr)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```
