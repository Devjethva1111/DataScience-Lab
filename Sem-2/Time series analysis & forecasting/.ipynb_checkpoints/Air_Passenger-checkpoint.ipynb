```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Unnamed: 0']
  ys = series['First.Class']
  
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Unnamed: 0', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Unnamed: 0')
_ = plt.ylabel('First.Class')
```

```python
import pandas as pd

# Assuming df is your DataFrame, and 'Economy.Class' is the column for which you want to calculate CMA
window_size = 3  # Adjust the window size as needed

# Calculate Central Moving Average
df['Economy_Class_CMA'] = df['Economy.Class'].rolling(window=window_size, center=True).mean()

# Plotting the original data and CMA
plt.figure(figsize=(10, 6))
plt.plot(df['Economy.Class'], label='Economy Class', color='#3498db', linewidth=2)
plt.plot(df['Economy_Class_CMA'], label=f'CMA {window_size}', linestyle='--', color='#e74c3c', linewidth=2)

# Adding labels and title
plt.title('Economy Class Central Moving Average', fontsize=16)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Adding legend with a better location
plt.legend(loc='upper left')

# Adding grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Adding a light gray background color for better contrast
plt.gca().set_facecolor('#ecf0f1')

# Show the plot
plt.show()
```

```python
import pandas as pd

# Assuming df is your DataFrame, and 'Economy.Class' is the column for which you want to calculate weighted CMA
window_size = 3  # Adjust the window size as needed
weights = [0.25, 0.5, 0.25]  # Adjust the weights according to your requirements

# Define a custom weighted average function
def weighted_average(x):
    return (x * weights).sum()

# Calculate Weighted Central Moving Average
df['Economy_Class_Weighted_CMA'] = df['Economy.Class'].rolling(window=window_size, center=True).apply(weighted_average, raw=True)

# Plotting the original data and Weighted CMA
plt.figure(figsize=(10, 6))
plt.plot(df['Economy.Class'], label='Economy Class', color='#3498db', linewidth=2)
plt.plot(df['Economy_Class_Weighted_CMA'], label=f'Weighted CMA', linestyle='--', color='#e74c3c', linewidth=2)

# Adding labels and title
plt.title('Economy Class Weighted Central Moving Average', fontsize=16)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Adding legend with a better location
plt.legend(loc='upper left')

# Adding grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Adding a light gray background color for better contrast
plt.gca().set_facecolor('#ecf0f1')

# Show the plot
plt.show()
```
