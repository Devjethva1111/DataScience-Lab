```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define ReLU function
def relu(x):
    return np.maximum(0, x)

# Generate x values
x = np.linspace(-5, 5,100)
```

```python
# Compute y values
y = relu(x)
```

```python
# Plot ReLU function
plt.plot(x, y, label='ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Function')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the non-convex function
def non_convex_function(x):
    return x**4 - 3*x**3 + 2

# Define the gradient of the non-convex function
def non_convex_gradient(x):
    return 4*x**3 - 9*x**2

# Gradient Descent function
def gradient_descent(gradient_func, initial_x, learning_rate, num_iterations):
    x = initial_x
    for _ in range(num_iterations):
        gradient = gradient_func(x)
        x -= learning_rate * gradient
    return x

# Generate x values
x_values = np.linspace(-5, 5, 100)
y_values = non_convex_function(x_values)

# Apply gradient descent to find the minimum
initial_x = 2.5  # Initial guess for minimum
learning_rate = 0.01
num_iterations = 100
min_x = gradient_descent(non_convex_gradient, initial_x, learning_rate, num_iterations)
min_y = non_convex_function(min_x)

# Plot the non-convex function and the minimum found by gradient descent
plt.plot(x_values, y_values, label='Non-Convex Function')
plt.scatter(min_x, min_y, color='red', label='Minimum Found by Gradient Descent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Non-Convex Function and Gradient Descent')
plt.grid(True)
plt.legend()
plt.show()
```
